```python
from collections import defaultdict

def relationship_between_columns(df, column1, column2):
  mapping = defaultdict(set)

  for index, row in df.iterrows():
      field1 = row[column1]
      field2 = row[column2]
      mapping[field1].add(field2)

  for field1, field2 in mapping.items():
      print(f"{column1}: {field1}, {column2}: {field2}")
```

```python

def check_campus_municipio(df):
  """
  Verifica se a segunda palavra no nome do campus corresponde ao município do campus.

  Args:
    df: O DataFrame contendo os dados.

  Returns:
    True se a segunda palavra no nome do campus sempre corresponder ao município do campus, False caso contrário.
  """
  for _, row in df.iterrows():
    campus_words = row['Campus'].split()
    if len(campus_words) > 1:
      segunda_palavra_campus = campus_words[1]
      if segunda_palavra_campus != row['Município do Campus']:
        return False
  return True

resultado = check_campus_municipio(data)
resultado
```

## Criando novo DF

```python
sanitized_data = pd.DataFrame()
sanitized_data['Evadidos_Flag'] = data['Evadidos_Flag']
sanitized_data['matricula_retida_neste_periodo'] = data['É matrícula retida neste período letivo?'].map({'Não Retida': 0, 'Retida': 1})
sanitized_data['eh_ingressante'] = data['É ingressante?'].map({'Não': 0, 'Sim': 1})
sanitized_data['matricula_retida'] = data['Matrícula retida?'].map({'Dentro do prazo': 0, 'Fora do prazo': 1})
sanitized_data['sexo'] = data['Sexo'].map({'F': 0, 'M': 1}) #

label_encoder = LabelEncoder()

mapping = {
    'Básico': 0,
    'Técnico': 1,
    'Graduação': 2,
    'Pós-graduação': 3
}

sanitized_data['nivel_ensino'] = data['Nível de Ensino'].map(mapping)
sanitized_data['cod_curso'] = label_encoder.fit_transform(data['Cód. do Curso'])

sanitized_data['ingresso'] = label_encoder.fit_transform(data['Forma de Ingresso'])

sanitized_data['periodo'] = data['Num Seq deste Período Letivo']
sanitized_data['ja_formou'] = data['ja_formou']
sanitized_data['periodos_cursados'] = data['Periodos Cursados']
sanitized_data['mora_onde_estuda'] = data['mora_onde_estuda']
sanitized_data['idade'] = data['idade']
sanitized_data['CH_normalizada'] = data['CH_normalizada']

sanitized_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

correlation_matrix = sanitized_data.corr()

target_corr = correlation_matrix['Evadidos_Flag'].abs()
ordered_vars = target_corr.sort_values(ascending=False).index

ordered_correlation_matrix = correlation_matrix.loc[ordered_vars, ordered_vars]

plt.figure(figsize=(12, 8))

sns.heatmap(
    ordered_correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

plt.title('Matriz de Correlação das Variáveis (Ordenada)', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

# Arvore de decisao



Detecção de Interações entre Variáveis: As árvores de decisão capturam automaticamente as interações entre as variáveis, o que pode ser importante nestecaso, onde diferentes combinações de fatores podem afetar a evasão.

- Problemas de classificação: Quando você deseja prever a classe de um novo exemplo
- Interpretabilidade é importante: Quando a compreensão das regras de decisão é crucial para o negócio.
- Dados com poucas características: Árvore de decisão pode ser eficiente com poucos atributos.


https://scikit-learn.org/stable/modules/tree.html

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

cols1 = ['periodos_cursados', 'cod_curso', 'periodo', 'ingresso', 'nivel_ensino', 'CH_normalizada']

X = sanitized_data[cols1]
y = sanitized_data['Evadidos_Flag']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

modelo = DecisionTreeClassifier(random_state=42)

modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia: {accuracy}")

cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusão:")
print(cm)

report = classification_report(y_test, y_pred)
print("Relatório de Classificação:")
print(report)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Visualizar a árvore
plt.figure(figsize=(40,20))
plot_tree(modelo, feature_names=cols1, class_names=['Não Evadidos', 'Evadidos'], filled=True)
plt.show()
```

```python
importances = modelo.feature_importances_
feature_names = cols1

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

print(importance_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

class_names = ['Não Evadidos', 'Evadidos']

plt.figure(figsize=(8, 6))

# Cria o heatmap
sns.heatmap(
    cm,
    annot=True,                  # Anota os valores na matriz
    fmt='d',                     # Formato dos valores anotados (d para inteiros)
    cmap='Blues',                # Mapa de cores para o heatmap
    xticklabels=class_names,     # Rótulos do eixo x
    yticklabels=class_names,     # Rótulos do eixo y
    linewidths=0.5,              # Largura das linhas que separam os blocos
    cbar_kws={'shrink': 0.8}     # Ajusta o tamanho da barra de cor
)

plt.title('Matriz de Confusão', fontsize=18)
plt.xlabel('Classe Predita', fontsize=14)
plt.ylabel('Classe Verdadeira', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
```

  - 46 (Falsos Positivos - FP): O modelo previu incorretamente que 49 estudantes evadiram, mas na verdade eles não evadiram6
- Segunda Linha (Evasão - 1):
   - 54 (Falsos Negativos - FN): O modelo previu incorretamente que 54 estudantes não evadiram, mas na verdade eles evadiram.
   - 212 (Verdadeiros Positivos - VP): O modelo previu corretamente que 241 estudantes evadiram.


Acurácia: indica uma performance geral do modelo. Dentre todas as classificações, quantas o modelo classificou corretamente; Neste modelo, em 77% dos casos

Precisão: dentre todas as classificações de classe Positivo que o modelo fez, quantas estão corretas;

Recall/Revocação/Sensibilidade: dentre todas as situações de classe Positivo como valor esperado, quantas estão corretas;

A precisão pode ser usada em uma situação em que os Falsos Positivos são considerados mais prejudiciais que os Falsos Negativos. Por exemplo, se for interessante que nosso modelo indique somente os alunos evadidos, mesmo que acabe classificando alguns que não evadiram como evadidos (situação de Falso Positivo) no processo. Ou seja, o modelo deve ser preciso em suas classificações, pois, a partir do momento que consideramos um aluno como evadido quando na verdade ele não é, poderíamos estar tomando decisões desnecessárias e gastando mais recursos em alunos que na verdade não iriam evadir.

O recall pode ser usado em uma situação em que os Falsos Negativos são considerados mais prejudiciais que os Falsos Positivos. Por exemplo, o modelo deve, de qualquer maneira, encontrar todos os alunos evadidos, mesmo que classifique alguns alunos que não evadiram como evadidos (situação de Falso Positivo) no processo. Ou seja, o modelo deve ter alto recall, pois classificar um aluno evadido como não evadido pode resultar na perda de oportunidades de intervenção para evitar que o aluno desista completamente.

Neste cenário, as taxas de precisão e recall foram muito próximas, o que resultou em um F1-Score igualmente próximo, indicando um bom equilíbrio entre identificar corretamente os alunos evadidos e minimizar os erros de classificação.
